## **Model** **Building**

```python
df.head(10)
```

```python
df1=df.drop(['Age_category'],axis=1)
```

```python
df1.head()
```

```python
df1.info()
```

```python
for feature in df1.columns:
    if df1[feature].dtype == 'object':
        print('\n')
        print('feature:',feature)
        print(pd.Categorical(df1[feature].unique()))
        print(pd.Categorical(df1[feature].unique()).codes)
        df1[feature] = pd.Categorical(df1[feature]).codes
```

```python
df1.head()
```

```python
X = df1.drop('satisfaction', axis = 1)
Y = df1['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.30, random_state=1,stratify=Y)
```

```python
print("X_train",len(X_train))
print("X_test",len(X_test))
print("Y_train",len(Y_train))
print("Y_test",len(Y_test))
```

```python
#Logistic model fit
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model.fit(X_train,Y_train)
```

```python
#predict Y
log_predict_y = log_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm= confusion_matrix(Y_test, log_predict_y)
cm
```

```python

sns.heatmap(cm,annot=True, fmt='d',cbar=False, cmap='rainbow')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(Y_test,log_predict_y))
```

```python
from sklearn.metrics import accuracy_score
log_accuracy_score = accuracy_score(Y_test, log_predict_y)
log_accuracy_score
```

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion = 'gini', random_state=1)
dt_model.fit(X_train,Y_train)
```

```python
dt_predict_y = dt_model.predict(X_test)
```

```python
cm1 = confusion_matrix(Y_test, dt_predict_y)
cm1
```

```python
sns.heatmap(cm1,annot=True, fmt='d',cbar=False, cmap='rainbow')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(Y_test,dt_predict_y))
```

```python
dt_accuracy_score = accuracy_score(Y_test, dt_predict_y)
dt_accuracy_score
```

```python
print (pd.DataFrame(dt_model.feature_importances_, columns = ["Imp"], index = X_train.columns).sort_values('Imp',ascending=False))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
param = {'criterion':['gini','entropy'],
         'max_depth':[20,22],
         'min_samples_split':[5,10],
         'min_samples_leaf':[3,5],
         'min_impurity_decrease':[0.0001,0.0002]}

grid_search_dt = GridSearchCV(estimator = dt_model, param_grid = param, cv = 10, n_jobs=-1,verbose=2)
grid_search_dt.fit(X_train,Y_train)
```

```python
grid_search_dt.best_params_
```
